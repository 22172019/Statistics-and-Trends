# Automobile Data Analysis Report


```python
# ----------------------- Imports -------------------------------------
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
from IPython.display import display, Markdown

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# ----------------------- 1. Load Dataset & EDA -----------------------
def load_data(file_path="/content/Automobile_data.csv"):
    """
    Load and clean the dataset. Handle missing values and outliers.
    Returns a cleaned dataframe.
    """
    display(Markdown("## 1. Dataset Loading & Initial Exploration"))
    display(Markdown("Loaded the automobile dataset and inspect its structure, types, and missing values."))
    df = pd.read_csv(file_path)
    df.replace('?', pd.NA, inplace=True)
    df.dropna(subset=['price', 'horsepower', 'fuel-type'], inplace=True)

    numeric_cols = ['price', 'horsepower', 'city-mpg', 'highway-mpg', 'engine-size', 'curb-weight']
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    df.dropna(subset=numeric_cols, inplace=True)

    # Handle outliers for price and horsepower
    for col in ['price','horsepower']:
        Q1, Q3 = df[col].quantile([0.25,0.75])
        IQR = Q3 - Q1
        df = df[(df[col] >= Q1 - 1.5*IQR) & (df[col] <= Q3 + 1.5*IQR)]

    # Display basic info
    display(Markdown("### 1.1 First 5 Rows"))
    display(df.head())

    display(Markdown("### 1.2 Dataset Info"))
    display(pd.DataFrame({"Column": df.columns, "Non-Null Count": df.notnull().sum(), "Data Type": df.dtypes}))

    display(Markdown("### 1.3 Summary Statistics"))
    display(df[numeric_cols].describe())

    display(Markdown("### 1.4 Categorical Value Counts"))
    for col in ['fuel-type', 'aspiration', 'body-style', 'drive-wheels']:
        if col in df.columns:
            display(Markdown(f"#### {col}"))
            display(df[col].value_counts().to_frame().rename(columns={col:'Count'}))

    display(Markdown("Dataset cleaned, missing values removed, and outliers handled for price and horsepower."))
    return df

# ----------------------- 2. Statistical Summary ----------------------
def statistical_summary(data):
    """
    Display descriptive statistics, 4 main moments, and correlation heatmap
    """
    numeric_cols = ['price','horsepower','city-mpg','highway-mpg','engine-size','curb-weight']
    display(Markdown("## 2. Descriptive Statistics & Moments"))

    # Summary stats
    display(Markdown("### 2.1 Summary Statistics"))
    display(data[numeric_cols].describe())

    # 4 main moments
    display(Markdown("### 2.2 Statistical Moments"))
    moments = pd.DataFrame({
        'Mean': data[numeric_cols].mean(),
        'Std': data[numeric_cols].std(),
        'Skew': data[numeric_cols].apply(skew),
        'Kurtosis': data[numeric_cols].apply(kurtosis)
    })
    display(moments)
    display(Markdown("Skew >0 indicates positive skew. Kurtosis >3 indicates heavy tails."))

    # Correlation heatmap
    display(Markdown("### 2.3 Correlation Heatmap"))
    corr = data[numeric_cols].corr()
    display(corr)
    plt.figure(figsize=(8,6))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', center=0, linewidths=1.5, linecolor='gray',
                cbar_kws={"shrink":0.8,"label":"Correlation Coefficient"}, annot_kws={"fontsize":11})
    plt.title("Correlation Heatmap", fontsize=16)
    plt.tight_layout()
    plt.show()

# ----------------------- 3. Relational Plots --------------------------
def plot_relational(data):
    """
    Create multiple relational plots with different features
    """
    display(Markdown("## 3. Relational Plots"))

    # Scatter: horsepower vs price
    display(Markdown("### 3.1 Scatter Plot: Horsepower vs Price"))
    plt.figure(figsize=(10,7))
    sns.scatterplot(data=data, x='horsepower', y='price', hue='fuel-type', style='fuel-type',
                    s=120, palette='Set2', edgecolor='black', alpha=0.85)
    top3 = data.nlargest(3,'price')
    for i,row in top3.iterrows():
        plt.text(row['horsepower']+2,row['price']+500,f"${int(row['price']):,}", color='red', fontweight='bold')
    plt.xlabel("Horsepower")
    plt.ylabel("Price")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

    # Line: city-mpg vs highway-mpg
    display(Markdown("### 3.2 Line Plot: City MPG vs Highway MPG"))
    plt.figure(figsize=(10,6))
    sorted_data = data.sort_values('city-mpg')
    plt.plot(sorted_data['city-mpg'], sorted_data['highway-mpg'], marker='o', color='green')
    plt.xlabel("City MPG")
    plt.ylabel("Highway MPG")
    plt.title("City MPG vs Highway MPG")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

    # Step: engine-size vs price
    display(Markdown("### 3.3 Step Plot: Engine Size vs Price"))
    plt.figure(figsize=(10,6))
    sorted_data = data.sort_values('engine-size')
    plt.step(sorted_data['engine-size'], sorted_data['price'], where='mid', color='purple')
    plt.xlabel("Engine Size")
    plt.ylabel("Price")
    plt.title("Engine Size vs Price")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# ----------------------- 4. Categorical Plots -------------------------
def plot_categorical(data):
    """
    Create multiple categorical plots with different features
    """
    display(Markdown("## 4. Categorical Plots"))

    # Bar plot: fuel-type vs avg price
    display(Markdown("### 4.1 Bar Plot: Fuel Type vs Average Price"))
    plt.figure(figsize=(9,6))
    avg_fuel = data.groupby('fuel-type')['price'].mean().sort_values()
    sns.barplot(x=avg_fuel.index, y=avg_fuel.values, palette='Set2', edgecolor='black')
    for i,v in enumerate(avg_fuel.values):
        plt.text(i, v+500, f"${int(v):,}", ha='center', fontweight='bold')
    plt.ylabel("Average Price")
    plt.xlabel("Fuel Type")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

    # Horizontal bar: body-style vs avg price
    display(Markdown("### 4.2 Horizontal Bar Plot: Body Style vs Average Price"))
    avg_body = data.groupby('body-style')['price'].mean().sort_values()
    plt.figure(figsize=(9,6))
    plt.barh(avg_body.index, avg_body.values, color='skyblue', edgecolor='black')
    for i,v in enumerate(avg_body.values):
        plt.text(v+500, i, f"${int(v):,}", va='center', fontweight='bold')
    plt.xlabel("Average Price")
    plt.ylabel("Body Style")
    plt.grid(axis='x', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

    # Histogram: price distribution
    display(Markdown("### 4.3 Histogram: Price Distribution"))
    plt.figure(figsize=(9,6))
    plt.hist(data['price'], bins=15, color='orange', edgecolor='black', alpha=0.7)
    plt.xlabel("Price")
    plt.ylabel("Count")
    plt.title("Price Distribution")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

    # Pie chart: drive-wheels distribution
    display(Markdown("### 4.4 Pie Chart: Drive Wheels Distribution"))
    drive_counts = data['drive-wheels'].value_counts()
    plt.figure(figsize=(7,7))
    plt.pie(drive_counts.values, labels=drive_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
    plt.title("Drive Wheels Proportion")
    plt.tight_layout()
    plt.show()

# ----------------------- 5. Statistical Plots ------------------------
def plot_statistical(data):
    """
    Create statistical plots using different features
    """
    display(Markdown("## 5. Statistical Plots"))

    # Boxplot: price by fuel-type
    display(Markdown("### 5.1 Boxplot: Price by Fuel Type"))
    plt.figure(figsize=(10,7))
    ax = sns.boxplot(x='fuel-type', y='price', data=data, palette='Set2', linewidth=1.5, showmeans=True,
                     meanprops={"marker":"o","markerfacecolor":"white","markeredgecolor":"black","markersize":10})
    sns.stripplot(x='fuel-type', y='price', data=data, color='black', size=4, jitter=True, alpha=0.6)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

    # Violin plot: price by body-style
    display(Markdown("### 5.2 Violin Plot: Price by Body Style"))
    plt.figure(figsize=(10,6))
    sns.violinplot(x='body-style', y='price', data=data, palette='Set2', inner='quartile')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

    # Pairplot: price, horsepower, engine-size, city-mpg
    display(Markdown("### 5.3 Pairplot: Price, Horsepower, Engine Size, City MPG"))
    sns.pairplot(data[['price','horsepower','engine-size','city-mpg']], diag_kind='kde', plot_kws={'alpha':0.6})
    plt.suptitle("Pairplot of Selected Numeric Features", fontsize=16, weight='bold', y=1.02)
    plt.show()

# ------------------ 6. Critical Assessment ----------------------------
def critical_assessment(data):
    display(Markdown("## 6. Critical Assessment Summary"))
    fuel_price = data.groupby('fuel-type')['price'].mean()
    horsepower_corr = data['horsepower'].corr(data['price'])
    skewness = data['price'].skew()
    summary_df = pd.DataFrame({
        'Observation': [
            "Fuel-type cars are more expensive on average",
            "Horsepower moderately correlates with price",
            "Price distribution is positively skewed"
        ],
        'Details': [
            f"Average price by fuel-type:\n{fuel_price.round(2)}",
            f"Correlation between horsepower and price: {horsepower_corr:.2f}",
            f"Skewness of price: {skewness:.2f}"
        ]
    })
    styled = summary_df.style.set_table_styles([
        {'selector': 'th', 'props': [('font-size', '16pt'), ('text-align', 'center')]},
        {'selector': 'td', 'props': [('font-size', '14pt'), ('text-align', 'left')]}
    ])
    display(styled)

# ------------------ 7. Key Recommendations ---------------------------
def key_recommendations(data):
    display(Markdown("## 7. Key Recommendations & Insights"))
    fuel_price = data.groupby('fuel-type')['price'].mean()
    horsepower_corr = data['horsepower'].corr(data['price'])
    price_skew = data['price'].skew()
    recommendations = [
        f"Fuel-type '{fuel_price.idxmax()}' cars are most expensive. Adjust marketing/pricing.",
        f"Horsepower moderately correlates with price ({horsepower_corr:.2f}). Performance-oriented cars may justify higher pricing.",
        f"Price is positively skewed (Skewness={price_skew:.2f}), indicating premium/high-end vehicles."
    ]
    for rec in recommendations:
        display(Markdown(f"- {rec}"))

# ------------------------- Main Function -----------------------------
def main():
    display(Markdown("# Automobile Data Analysis Report"))
    df = load_data()
    statistical_summary(df)
    plot_relational(df)
    plot_categorical(df)
    plot_statistical(df)
    critical_assessment(df)
    key_recommendations(df)

# ---------------------- Script Entry Point ---------------------------
if __name__ == "__main__":
    main()
```
